```python
from fbprophet import Prophet
import json
from fbprophet.serialize import model_to_json, model_from_json
with open("../models/serialized_prophet_model.json", 'r') as fin:
    model_Loaded = model_from_json(json.load(fin))  # Load model
```

```python
future = model_Loaded.make_future_dataframe(periods=7*50)  
forecast = model_Loaded.predict(future)
forecast[['ds','trend', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
model_Loaded.plot(forecast); # visualise the result
```
